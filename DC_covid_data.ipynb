```python
import pandas as pd
import numpy as np
import flask
import pymongo
import json
from datetime import datetime
```

```python
covid_data = 'july-31-dc-covid.csv'
```

```python
dc_covid_df = pd.read_csv(covid_data)
```

```python
dc_covid_df = dc_covid_df.sort_values('Date', ascending = False)
```

```python
dc_covid_df
```

```python
json_dc_covid = dc_covid_df.to_json(orient = 'records')
parsed = json.loads(json_dc_covid)
print(json.dumps(parsed, indent = 4, sort_keys= True))
```

```python
with open('dc_covid_july31.json', 'w') as f:
    json.dump(parsed, f, indent=4)
```
